```python
import pandas as pd
import scipy.stats as stats

# 분석 대상 컬럼
cols = ['차종류','연령','운전자','경력','ABS','AIRBAG','MOZEN','위반','성별']

significant_vars = []

for col in cols:
    # 각 그룹별 cha_los2 리스트 만들기
    groups = [group['cha_los2'].values for name, group in df.groupby(col)]

    # 그룹이 2개 이상이어야 ANOVA 가능
    if len(groups) < 2:
        continue

    # ANOVA 수행
    f_stat, p_val = stats.f_oneway(*groups)

    print(f"변수: {col}")
    print(f"  F 통계량: {f_stat:.4f}")
    print(f"  p-value: {p_val:.6f}")

    # 유의수준 0.05 기준
    if p_val < 0.05:
        significant_vars.append(col)

    print("-" * 40)

print("\n유의한 변수들:")
print(significant_vars)
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

# Cramer's V 계산 함수
def cramers_v(x, y):
    table = pd.crosstab(x, y)
    chi2 = stats.chi2_contingency(table)[0]
    n = table.sum().sum()
    phi2 = chi2 / n
    r, k = table.shape
    return np.sqrt(phi2 / min(k - 1, r - 1))

# 확인할 변수
cols = ['ABS', 'AIRBAG', 'MOZEN']

# 상관계수 매트릭스 생성
corr_mat = pd.DataFrame(index=cols, columns=cols, dtype=float)

for c1 in cols:
    for c2 in cols:
        corr_mat.loc[c1, c2] = cramers_v(df[c1], df[c2])

print(corr_mat)
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

def cramers_v_with_stats(x, y):
    # 교차표
    table = pd.crosstab(x, y)

    # 카이제곱 검정
    chi2, p, dof, expected = stats.chi2_contingency(table)

    # Cramer's V 계산
    n = table.sum().sum()
    phi2 = chi2 / n
    r, k = table.shape
    cramers_v = np.sqrt(phi2 / min(k - 1, r - 1))

    return cramers_v, chi2, p

# 확인할 변수들
cols = ['ABS', 'AIRBAG', 'MOZEN']

# 결과 저장용 데이터프레임
results = []

for c1 in cols:
    for c2 in cols:
        v, chi2, p = cramers_v_with_stats(df[c1], df[c2])
        results.append([c1, c2, v, chi2, p])

# 보기 좋게 정리
result_df = pd.DataFrame(results,
                         columns=['Var1', 'Var2', 'CramersV', 'Chi2', 'p_value'])

print(result_df)
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

# Cramer's V + p-value 함수
def cramers_v_with_p(x, y):
    table = pd.crosstab(x, y)
    chi2, p, dof, expected = stats.chi2_contingency(table)
    n = table.sum().sum()
    phi2 = chi2 / n
    r, k = table.shape
    v = np.sqrt(phi2 / min(k - 1, r - 1))
    return v, p

# 분석할 변수 목록
cols = ['ABS', 'AIRBAG', 'MOZEN']

# 결과 저장용 DF 생성
v_mat = pd.DataFrame(index=cols, columns=cols, dtype=float)
p_mat = pd.DataFrame(index=cols, columns=cols, dtype=float)

for c1 in cols:
    for c2 in cols:
        v, p = cramers_v_with_p(df[c1], df[c2])
        v_mat.loc[c1, c2] = v
        p_mat.loc[c1, c2] = p

print("=== Cramer's V Matrix ===")
print(v_mat)

print("\n=== Chi-square p-value Matrix ===")
print(p_mat)
```

```python
# 1. 각 요율요소별 Risk 계산 함수
def calculate_relative_risk(data, factor_col, loss_col='cha_los2'):
    """
    요율요소별 상대도 계산
    - 건수, 손해액 합계
    - Risk = 평균 손해액
    - 상대도 = Risk / 최소 Risk (가장 안전한 그룹을 1로 설정)
    """
    # 그룹별 집계
    grouped = data.groupby(factor_col).agg({
        loss_col: ['count', 'sum', 'mean']
    }).reset_index()

    # 컬럼명 정리
    grouped.columns = [factor_col, '건수', '총손해액', '평균손해액(Risk)']

    # 상대도 계산: 최소 Risk를 1로 설정
    min_risk = grouped['평균손해액(Risk)'].min()
    grouped['상대도'] = grouped['평균손해액(Risk)'] / min_risk

    # 정렬
    grouped = grouped.sort_values(factor_col)

    return grouped

# 분석할 요율요소 목록
risk_factors = ['차종류', '연령', '운전자', '경력', 'ABS', 'AIRBAG', 'MOZEN', '위반', '성별']

print("="*80)
print("요율요소별 상대도 분석 결과")
print("="*80)
```

```python
# 차종류별 차량 담보 심도(cha_los2) 분석
차종류_상대도 = calculate_relative_risk(df, '차종류', loss_col='cha_los2')
print("\n[차종류별 차량담보 심도 상대도 분석]")
print("="*70)
print(차종류_상대도.to_string(index=False))
print("="*70)
print(f"\n기준그룹(상대도=1.0): 차종류 그룹 {차종류_상대도.loc[차종류_상대도['상대도'].idxmin(), '차종류']}")
print(f"최소 Risk(평균손해액): {차종류_상대도['평균손해액(Risk)'].min():.2f}원")
print(f"\n상대도 범위: {차종류_상대도['상대도'].min():.3f} ~ {차종류_상대도['상대도'].max():.3f}")

# 테이블 저장
차종류_상대도
```

```python
# 1. 각 요율요소별 Risk 계산 함수
def calculate_relative_risk(data, factor_col, loss_col='cha_los2'):
    """
    요율요소별 상대도 계산
    - 건수, 손해액 합계
    - Risk = 평균 손해액
    - 상대도 = Risk / 최소 Risk (가장 안전한 그룹을 1로 설정)
    """
    # 그룹별 집계
    grouped = data.groupby(factor_col).agg({
        loss_col: ['count', 'sum', 'mean']
    }).reset_index()

    # 컬럼명 정리
    grouped.columns = [factor_col, '건수', '총손해액', '평균손해액(Risk)']

    # 상대도 계산: 최소 Risk를 1로 설정
    min_risk = grouped['평균손해액(Risk)'].min()
    grouped['상대도'] = grouped['평균손해액(Risk)'] / min_risk

    # 정렬
    grouped = grouped.sort_values(factor_col)

    return grouped

# 분석할 요율요소 목록
risk_factors = ['차종류', '연령', '운전자', '경력', 'ABS', 'AIRBAG', 'MOZEN', '위반', '성별']

print("="*80)
print("요율요소별 상대도 분석 결과")
print("="*80)
```

```python
import pandas as pd
import numpy as np
import ipywidgets as widgets
from IPython.display import display, HTML
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm

class InsurancePremiumCalculator:
    def __init__(self, data_path):
        """보험료 계산기 초기화"""
        # 데이터 로드
        self.df = pd.read_csv(data_path, encoding='cp949')
        
        # 상대도 컬럼 목록
        self.relative_columns = [col for col in self.df.columns if col.endswith('_상대도')]
        
        # 각 속성별 가능한 값과 상대도 매핑 생성
        self.factor_mapping = self._create_factor_mapping()
        
    def _create_factor_mapping(self):
        """각 속성별로 가능한 값과 해당 상대도를 매핑하는 딕셔너리 생성"""
        mapping = {}
        
        # 1. 차종류 (car_knd) 매핑
        car_mapping = {
            '1': 1.0,    # 소형a
            '2': 1.109,  # 소형b
            '3': 1.109,  # 중형
            '4': 1.455,  # 대형
            '5': 1.0     # 다목적
        }
        mapping['차종류'] = car_mapping
        
        # 2. 운전자 (DRV_TNK) 매핑
        driver_mapping = {
            '1': 1.0,    # 1인
            '2': 1.0,    # 부부
            '3': 1.0,    # 가족
            '4': 1.0,    # 가족확대
            '5': 1.0,    # 가족확대
            '0': 1.361   # 누구나
        }
        mapping['운전자'] = driver_mapping
        
        # 3. MOZEN (moz_yn) 매핑
        mozen_mapping = {
            '1': 1.0,    # 안전장치있음
            '2': 1.092,  # 안전장치없음
            '안전장치 있음': 1.0,
            '안전장치 없음': 1.092
        }
        mapping['MOZEN'] = mozen_mapping
        
        # 4. 위반 (VIO_YUL) 매핑
        violation_mapping = {
            '1': 1.0,    # 장기무위반
            '2': 1.168,  # 무위반
            '3': 1.168   # 위반
        }
        mapping['위반'] = violation_mapping
        
        # 5. 경력 (GAI_COD) 매핑
        career_mapping = {
            '1': 1.0,    # 능숙
            '2': 1.131,  # 중급
            '3': 1.540   # 초보
        }
        mapping['경력구간'] = career_mapping
        
        # 6. 나이별 상대도 (1살 단위)
        age_ratios = {
            20: 2.496002, 21: 2.404186, 22: 2.315277, 23: 2.229275, 24: 2.146181,
            25: 2.065995, 26: 1.988716, 27: 1.914345, 28: 1.842881, 29: 1.774324,
            30: 1.708675, 31: 1.645934, 32: 1.586100, 33: 1.529174, 34: 1.475155,
            35: 1.424044, 36: 1.375840, 37: 1.330543, 38: 1.288155, 39: 1.248673,
            40: 1.212100, 41: 1.178433, 42: 1.147674, 43: 1.119823, 44: 1.094879,
            45: 1.072843, 46: 1.053715, 47: 1.037493, 48: 1.024180, 49: 1.013773,
            50: 1.006275, 51: 1.001684, 52: 1.0, 53: 1.001224, 54: 1.005355,
            55: 1.012394, 56: 1.022341, 57: 1.035194, 58: 1.050956, 59: 1.069625,
            60: 1.091201, 61: 1.115685, 62: 1.143077, 63: 1.173376, 64: 1.206582,
            65: 1.242696, 66: 1.281718, 67: 1.323647, 68: 1.368483, 69: 1.416227,
            70: 1.466879, 71: 1.520438, 72: 1.576904, 73: 1.636278, 74: 1.698560,
            75: 1.763749, 76: 1.831846, 77: 1.902850, 78: 1.976762, 79: 2.053581,
            80: 2.133307, 81: 2.215942, 82: 2.301483
        }
        
        # 18~19세는 20세와 동일하게 설정
        age_mapping = {age: age_ratios[20] for age in range(18, 20)}
        # 20~82세는 실제 값 사용
        age_mapping.update(age_ratios)
        # 83세 이상은 82세와 동일하게 설정
        max_age = max(age_ratios.keys())
        max_ratio = age_ratios[max_age]
        for age in range(max_age + 1, 101):  # 100세까지
            age_mapping[age] = max_ratio
        
        mapping['나이'] = age_mapping  # '나이대' 대신 '나이' 키 사용
        
        return mapping
    
    def _get_age_ratio(self, age):
        """나이에 해당하는 상대도 반환"""
        age = int(age)
        if age in self.factor_mapping['나이']:
            return self.factor_mapping['나이'][age]
        # 범위를 벗어나는 경우 가장 가까운 값 반환
        if age < 18:
            return self.factor_mapping['나이'][18]
        return self.factor_mapping['나이'][max(self.factor_mapping['나이'].keys())]
    
    def _get_violation_group(self, violation):
        """위반 점수를 그룹으로 변환"""
        violation = int(violation)
        if violation == -0.59:
            return '1'  # 장기무위반
        elif violation == 0:
            return '2'  # 무위반
        else:
            return '3'  # 위반
    
    def _get_career_group(self, years):
        """경력 기간을 그룹으로 변환"""
        years = int(years)
        if years <= 1:
            return '3'  # 초보
        elif years <= 5:
            return '2'  # 중급
        else:
            return '1'  # 능숙
    
    def calculate_premium(self, base_premium, factors):
        """
        보험료 계산
        
        Parameters:
        - base_premium: 기준 보험료
        - factors: 사전 형태의 인자들 (차종류, 나이, 운전자, MOZEN, 위반점수, 경력)
        
        Returns:
        - 계산된 보험료
        """
        # 기본 상대도 (1.0으로 초기화)
        ratios = {
            '차종류_상대도': 1.0,
            '나이_상대도': 1.0,
            '운전자_상대도': 1.0,
            'MOZEN_상대도': 1.0,
            '위반_상대도': 1.0,
            '경력_상대도': 1.0
        }
        
        try:
            # 1. 차종류 (1~5)
            if '차종류' in factors:
                car_type = str(factors['차종류']).strip()
                if car_type in self.factor_mapping['차종류']:
                    ratios['차종류_상대도'] = float(self.factor_mapping['차종류'][car_type])
            
            # 2. 나이 (1살 단위)
            if '나이' in factors:
                age = int(factors['나이'])
                ratios['나이_상대도'] = float(self._get_age_ratio(age))
            
            # 3. 운전자 유형 (0~5)
            if '운전자' in factors:
                driver_type = str(factors['운전자']).strip()
                if driver_type in self.factor_mapping['운전자']:
                    ratios['운전자_상대도'] = float(self.factor_mapping['운전자'][driver_type])
            
            # 4. MOZEN (안전장치)
            if 'MOZEN' in factors:
                mozen = str(factors['MOZEN']).strip()
                if mozen in self.factor_mapping['MOZEN']:
                    ratios['MOZEN_상대도'] = float(self.factor_mapping['MOZEN'][mozen])
            
            # 5. 위반 점수
            if '위반점수' in factors:
                violation = int(factors['위반점수'])
                violation_group = self._get_violation_group(violation)
                if violation_group in self.factor_mapping['위반']:
                    ratios['위반_상대도'] = float(self.factor_mapping['위반'][violation_group])
            
            # 6. 경력
            if '경력' in factors:
                career = int(factors['경력'])
                career_group = self._get_career_group(career)
                if career_group in self.factor_mapping['경력구간']:
                    ratios['경력_상대도'] = float(self.factor_mapping['경력구간'][career_group])
            
            # nan 값 확인 및 처리
            for key in ratios:
                if pd.isna(ratios[key]):
                    ratios[key] = 1.0
                    print(f"경고: {key}에 대한 상대도가 없습니다. 기본값 1.0을 사용합니다.")
            
            # 보험료 계산
            premium = base_premium
            for ratio in ratios.values():
                premium *= ratio
            
            # Off-balance 계수 적용 (38.3% 감소)
            off_balance_ratio = 0.617  # 1 - 0.383 (38.3% 감소)
            
            final_premium = premium * off_balance_ratio
            
            return {
                '기준 보험료': base_premium,
                '적용된 상대도': ratios,
                '상대도 적용 보험료': premium,
                'Off-balance 계수': off_balance_ratio,
                '최종 보험료': final_premium
            }
        
        except Exception as e:
            print(f"보험료 계산 중 오류가 발생했습니다: {str(e)}")
            import traceback
            traceback.print_exc()
            return {
                '기준 보험료': base_premium,
                '적용된 상대도': ratios,
                '상대도 적용 보험료': float('nan'),
                'Off-balance 계수': 0.617,
                '최종 보험료': float('nan'),
                '오류': str(e)
            }

# 계산기 인스턴스 생성
calculator = InsurancePremiumCalculator('c:/Users/THKIM/Desktop/3학년 2학기/금융데이터 분석/df_processed.csv')

# 스타일 설정
style = {'description_width': '120px'}
layout = widgets.Layout(width='400px')

# 1. 차종류 (1~5)
차종류_위젯 = widgets.Dropdown(
    options=[('1 (소형a)', '1'), ('2 (소형b)', '2'), ('3 (중형)', '3'), 
             ('4 (대형)', '4'), ('5 (다목적)', '5')],
    value='1',
    description='🚗 차종류:',
    style={'description_width': '80px'}
)

# 2. 나이 (18~100세)
나이_위젯 = widgets.IntSlider(
    value=35,
    min=18,
    max=100,
    step=1,
    description='🎂 나이:',
    continuous_update=True,
    style={'description_width': '80px'},
    layout=widgets.Layout(width='400px')
)

# 3. 운전자 (0,1,2,3,4,5)
운전자_위젯 = widgets.Dropdown(
    options=[('1인', '1'), ('부부', '2'), ('가족', '3'), 
            ('가족확대', '4'), ('가족확대', '5'), ('누구나', '0')],
    value='1',
    description='👨‍👩‍👧‍👦 운전자:',
    style={'description_width': '80px'}
)

# 4. MOZEN (안전장치)
MOZEN_위젯 = widgets.Dropdown(
    options=[('안전장치 있음', '안전장치 있음'), 
             ('안전장치 없음', '안전장치 없음')],
    value='안전장치 있음',
    description='🔒 안전장치:',
    style={'description_width': '80px'}
)

# 5. 위반점수
위반점수_위젯 = widgets.Dropdown(
    options=[('장기무위반 (1.0배)', -0.59), 
             ('무위반 (1.168배)', 0), 
             ('위반 (1.168배)', 1)],
    value=0,
    description='🚨 위반내역:',
    style={'description_width': '80px'}
)

# 6. 경력 (0~50년)
경력_위젯 = widgets.IntSlider(
    value=5,
    min=0,
    max=50,
    step=1,
    description='📅 운전경력:',
    style={'description_width': '80px'},
    layout=widgets.Layout(width='400px')
)

# 7. 기준 보험료
기준보험료_위젯 = widgets.FloatText(
    value=1000000,
    description='💰 기준 보험료:',
    style={'description_width': '100px'},
    layout=widgets.Layout(width='200px')
)

# 계산 버튼
계산_버튼 = widgets.Button(
    description='보험료 계산하기',
    button_style='success',
    icon='calculator',
    layout=widgets.Layout(width='300px', height='40px', margin='20px 0 0 0')
)

# 결과 출력 영역
결과_출력 = widgets.Output()

# UI 레이아웃
입력_박스 = widgets.VBox([
    widgets.HTML("<h2>📝 보험료 계산기</h2>"),
    widgets.HTML("<p>아래 입력란을 채우고 계산 버튼을 눌러주세요</p>"),
    차종류_위젯,
    나이_위젯,
    운전자_위젯,
    MOZEN_위젯,
    위반점수_위젯,
    경력_위젯,
    기준보험료_위젯,
    widgets.HBox([widgets.HTML('<div style="width:100px"></div>'), 계산_버튼]),
    결과_출력
], layout=widgets.Layout(
    width='450px',
    padding='20px',
    border='1px solid #e0e0e0',
    border_radius='5px',
    box_shadow='2px 2px 5px rgba(0,0,0,0.1)'
))

def calculate_premium_ui(button):
    with 결과_출력:
        결과_출력.clear_output(wait=True)
        try:
            factors = {
                '차종류': str(차종류_위젯.value),
                '나이': int(나이_위젯.value),
                '운전자': str(운전자_위젯.value),
                'MOZEN': MOZEN_위젯.value,
                '위반점수': float(위반점수_위젯.value),
                '경력': int(경력_위젯.value)
            }
            result = calculator.calculate_premium(float(기준보험료_위젯.value), factors)
            
            # 결과 스타일
            display(HTML(f"""
            <style>
                .result-box {{
                    margin-top: 20px;
                    padding: 15px;
                    background-color: #f8f9fa;
                    border-radius: 5px;
                    border-left: 5px solid #4e73df;
                }}
                .result-title {{
                    color: #2e59d9;
                    font-size: 1.2em;
                    font-weight: bold;
                    margin-bottom: 10px;
                }}
                .result-table {{
                    width: 100%;
                    border-collapse: collapse;
                    margin-top: 10px;
                }}
                .result-table td {{
                    padding: 8px 5px;
                    border-bottom: 1px solid #eee;
                }}
                .result-table tr:last-child td {{
                    border-bottom: none;
                }}
                .highlight {{
                    color: #2e59d9;
                    font-weight: bold;
                    font-size: 1.1em;
                }}
            </style>
            
            <div class="result-box">
                <div class="result-title">📊 보험료 계산 결과</div>
                <table class="result-table">
                    <tr>
                        <td>기준 보험료:</td>
                        <td style="text-align:right;">{result['기준 보험료']:,.0f}원</td>
                    </tr>
                    <tr>
                        <td colspan="2"><b>적용된 상대도:</b></td>
                    </tr>
            """))
            
            for factor, ratio in result['적용된 상대도'].items():
                display(HTML(f"""
                    <tr>
                        <td style="padding-left:20px;">- {factor}:</td>
                        <td style="text-align:right;">{ratio:.4f}</td>
                    </tr>
                """))
            
            display(HTML(f"""
                    <tr>
                        <td>상대도 적용 보험료:</td>
                        <td style="text-align:right;">{result['상대도 적용 보험료']:,.0f}원</td>
                    </tr>
                    <tr>
                        <td>Off-balance 계수 (38.3% 감소):</td>
                        <td style="text-align:right;">{result['Off-balance 계수']:.6f}</td>
                    </tr>
                    <tr>
                        <td><b>최종 보험료:</b></td>
                        <td style="text-align:right;"><span class="highlight">{result['최종 보험료']:,.0f}원</span></td>
                    </tr>
                </table>
            </div>
            """))
            
        except Exception as e:
            display(HTML("""
                <div style="color: red; background-color: #ffebee; padding: 10px; margin-top: 10px;">
                    ⚠️ 오류가 발생했습니다. 입력값을 확인해주세요.
                </div>
            """))
            print(f"오류 상세: {str(e)}")

# 버튼 클릭 이벤트 연결
계산_버튼.on_click(calculate_premium_ui)

# UI 표시
display(widgets.VBox([입력_박스], 
    layout=widgets.Layout(
        display='flex',
        flex_flow='column',
        align_items='center',
        width='100%'
    )
))
```
